```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("bitcoin.csv")
df.head()
```

```python

# Convert 'timestamp' to datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Calculate daily returns
df['Daily_Return'] = df['prices'].pct_change()
df = df.dropna()  # Drop the first row with NaN

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 10))
axes[0].plot(df['timestamp'], df['prices'], marker='o', linestyle='-')
axes[0].set_title('Close Price Over Time')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Close Price')
axes[0].grid(True)

axes[1].plot(df['timestamp'], df['Daily_Return'], marker='o', linestyle='-', color='r')
axes[1].set_title('Daily Return Over Time')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Daily Return')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Split data into training and test sets
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]
```

```python

# ARIMA model
model = ARIMA(abs(train['Daily_Return']), order=(1, 0, 0))
model_fit = model.fit()
model_fit.summary()

# Forecasting on test data
n_forecast = len(test)
forecast = model_fit.forecast(steps=n_forecast)

# Evaluate ARIMA model
rmse_arima = np.sqrt(mean_squared_error(test['Daily_Return'], forecast))
print(f'ARIMA RMSE: {rmse_arima}')
```

```python

# GARCH model
garch_model = arch_model(train['Daily_Return'], vol='Garch', p=1, q=1, mean='Zero')
garch_fit = garch_model.fit(update_freq=5)
garch_fit.summary()

# Forecasting volatility on test data
forecasts = garch_fit.forecast(start=0, horizon=n_forecast-1) # Subtract 1 because of zero-indexing
forecasted_variance = forecasts.variance.iloc[-1]
forecasted_volatility = np.sqrt(forecasted_variance)

# Make sure the forecasted_volatility length matches with the test set
if len(forecasted_volatility) != len(test):
    test = test.iloc[:-1]  # Remove the last entry to match the forecast length

    
# Calculate rolling standard deviation for the test set with a reasonable window, say 10 days
test['Rolling_Std'] = test['Daily_Return'].rolling(window=20).std()

# Drop NaN values generated by the rolling window calculation
test = test.dropna()

# Adjust your forecasted_volatility array to match this length
forecasted_volatility_adjusted = forecasted_volatility[-len(test):]

# Calculate RMSE
rmse_garch = np.sqrt(mean_squared_error(test['Rolling_Std'], forecasted_volatility_adjusted))
print(f'GARCH RMSE: {rmse_garch}')
```

```python
# Calculate rolling volatility using a 20-day window
window_size = 20
sma_volatility = test['Daily_Return'].rolling(window=window_size).std()

# First drop NaN values from both Series
sma_volatility = sma_volatility.dropna()
rolling_std = test['Rolling_Std'].dropna()

# Make sure they have the same length
if len(sma_volatility) != len(rolling_std):
    min_length = min(len(sma_volatility), len(rolling_std))
    sma_volatility = sma_volatility[-min_length:]
    rolling_std = rolling_std[-min_length:]

# Now you can safely calculate RMSE
rmse_sma = np.sqrt(mean_squared_error(sma_volatility, rolling_std))
rmse_sma
```

```python
# Code snippet to add EWMA to the existing framework
# This assumes `test['Daily_Return']` contains the daily returns in the test set

# Define the decay factor for EWMA
decay_factor = 0.94

# Calculate EWMA volatility
ewma_volatility = test['Daily_Return'].ewm(span=(2/(1-decay_factor)-1)).std()

# Drop NaN values
ewma_volatility = ewma_volatility.dropna()

# Make sure the forecasted_volatility length matches with the test set
if len(forecasted_volatility) != len(test):
    test = test.iloc[:-1]  # Remove the last entry to match the forecast length

# Evaluate EWMA model
rmse_ewma = np.sqrt(mean_squared_error(ewma_volatility, forecasted_volatility[:len(ewma_volatility)]))
print(f'EWMA RMSE: {rmse_ewma}')

# Now you can compare the RMSE of EWMA with GARCH and other models
```
